```python
%%capture --no-display
from pytrends.request import  TrendReq

conn = TrendReq(hl='en-US', tz=360, requests_args={'verify':False}, retries=2)
kw_list = ['Innovation', 'Technology']
conn.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='', gprop='')
related_words = conn.related_queries()
```

```python
%%capture --no-display
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = stopwords.words('english')
from pandas import DataFrame
import matplotlib.pyplot as plt
def plot_pie_chart(data:DataFrame, title):
    chart_data = {"word": [], "count": []}
    for index, row in data.iterrows():
        line = row['query'].split(" ")
        for word in line:
            if word not in stop_words and word.lower() != title.lower(): 
                if word in chart_data['word']:
                    chart_data['count'][chart_data['word'].index(word)] += row['value']
                else:
                    chart_data['word'].append(word)
                    chart_data['count'].append(row['value'])
    plt.title(title)
    plt.pie(chart_data['count'], labels=chart_data['word'], autopct='%1.1f%%')
plt.figure(figsize=(20,20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.subplot(2, 2, 1)
plot_pie_chart(related_words['Innovation']['top'], 'Innovation')
plt.subplot(2, 2, 2)
plot_pie_chart(related_words['Technology']['top'], 'Technology')

plt.subplot(2, 2, 3)
plot_pie_chart(related_words['Innovation']['rising'], 'Innovation')
plt.subplot(2, 2, 4)
plot_pie_chart(related_words['Technology']['rising'], 'Technology')
```
